```python
import pandas as pd

import numpy as np

from sklearn.model_selection import train_test_split

#modelimizdeki bağımlı değişkenin numeric halde yazılabilmesi için labelencoder'i kullanıyoruz ama
#bağımsız değişkenleri numeric hale getirirken ise one hot encoder'i kullanıyoruz
from sklearn.preprocessing import LabelEncoder

#burada bağımsız değişkenleri düzenlerken one hot encoderi kullanmalıyız
from sklearn.preprocessing import OneHotEncoder

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df.head(10)
```

```python
# bağımlı ve bağımsız değişkenlerimizi ayarlıyoruz ancak çok önemli bi nokta var ben her şeyin object türünde olmasını istiyorum ki bağımlı değişken için label encoder ya da bağımsız değişkenler için one hot encoder yaparken
#tip sorunu yaşamayayım, o yüzden direkt olarak bunları object yapıyorum yani str

# bağımlı değişkenimizi ayarlayalım
y = df['class'].astype(str)

#bağımsız değişkenimizi ayarlayalım
x = df.drop(columns = ['class']).astype(str)
```

```python
# datasetimizi train ve test olmak üzere bölüyoruz
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
df.info()
```

```python
#modelimizdeki bağımlı değişkenin numeric halde yazılabilmesi için labelencoder'i kullanıyoruz ama
#bağımsız değişkenleri numeric hale getirirken ise one hot encoder'i kullanıyoruz

#bağımlı değişkenimizi label encoder ile düzenleyelim
le = LabelEncoder()

#Train değişkeninde fit ve transform yapıyoruz
y_train = le.fit_transform(y_train)

#Keras float sever o yüzden böyle yapmalıyız
y_train = y_train.astype(np.float32)

#Test değişkeninde sadece transform yapıyoruz
y_test = le.transform(y_test)

#Keras float sever o yüzden böyle yapmalıyız
y_test = y_test.astype(np.float32)
```

```python
#modelimizdeki bağımlı değişkenin numeric halde yazılabilmesi için labelencoder'i kullanıyoruz ama
#bağımsız değişkenleri numeric hale getirirken ise one hot encoder'i kullanıyoruz

# bağımsız değişkenimizi one hot encoder ile düzenleyelim
ohe = OneHotEncoder(sparse_output=False)

#Train değişkeninde fit ve transform yapıyoruz
x_train = ohe.fit_transform(x_train).astype(np.float32)

#Keras float sever o yüzden böyle yapmalıyız, bu şekilde de yapılabilinir yani
#x_train = x_train.astype(np.float32)

#Test değişkeninde sadece transform yapıyoruz
x_test = ohe.transform(x_test).astype(np.float32)

#Keras float sever o yüzden böyle yapmalıyız, bu şekilde de yapılabilinir yani
#x_test = x_test.astype(np.float32)
```

```python
# modelimizi çapıralım
model = Sequential()

# kaç adet input olduğunu otomatik olarak bulalım
dim_of_input = x_train.shape[1]

# input layeri ayarlayalım, burada input_dim'i bulmalıyız onu bulalım
model.add(Dense(32,input_dim=dim_of_input))

# hidden layeri ayarlayalım
model.add(Dense(32,activation = 'relu'))
model.add(Dense(64,activation = 'relu'))
model.add(Dense(32,activation = 'relu'))

# output layeri ayarlayalım, output için sigmoid fonksiyonu kullanacağız yani ya 0 ya 1 veren relu falan kullanılamaz burada
# sigmoid activation classification için outputta verilir
model.add(Dense(1,activation = 'sigmoid'))

# sistemin nasıl çalışacağını ayarlayalım
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])

# modelimize öğretelim, epochs veriler ağda kaç kere dönsün, batch_size da verileri kaçar kaçar yollayacağımızı ayarlamamızı sağlıyor
# normalde testler öğrenme bittikten sonra yapılır ama biz burada validation_split yaparak burada öğrenirken test ediyoruz yani burada öğrenme daha güçlü oluyor
model.fit(x_train, y_train, epochs=64, batch_size=128, validation_split=0.2)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
